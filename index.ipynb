```python
import pandas as pd

# Baca data dari CSV file
dataset = pd.read_csv('KRB_Grid_SB250.csv', usecols=range(0, 13), header=1)

# Rule1. Rawan Bencana Tinggi (awal) setiap grid (n), RBT (1) = 0.00
# Rule 2. Jika resiko banjir tinggi V1(n) ≥ 20 maka Rule 2
def brt_n_v1(row):
  if row['BNJRTING'] < 20:
    return 0
  else:
    return row['BNJRTING']

# Tambah kolom BRT(n) V1 berdasarkan hasil rule 1 dan 2
dataset['BRT (n) V1'] = dataset.apply(lambda row: brt_n_v1(row), axis=1)

# Rule 3. Jika resiko banjir bandang tinggi V3(n) ≥ 10 maka Rule 3
def brt_n_v3(row):
  if row['BBNDGTING'] < 10:
    return 0
  else:
    return row['BBNDGTING']

# Tambah kolom BRT(n) V1 berdasarkan hasil rule 1 dan 2
dataset['BRT (n) V3'] = dataset.apply(lambda row: brt_n_v3(row), axis=1)

# Lihat beberapa baris pertama, sekedar mengecek
dataset.head()
```
