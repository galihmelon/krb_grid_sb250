```python
import pandas as pd

# Baca data dari CSV file
dataset = pd.read_csv('KRB_Grid_SB250.csv', usecols=range(0, 13), header=1)

# Rule1. Rawan Bencana Tinggi (awal) setiap grid (n), RBT (1) = 0.00
# Rule 2. Jika resiko banjir tinggi V1(n) ≥ 20
def brt_n_v1(row):
  if row['BNJRTING'] < 20:
    return 0
  else:
    return row['BNJRTING']

# Tambah kolom BRT(n) V1 berdasarkan hasil rule 1 dan 2
dataset['BRT(n) V1'] = dataset.apply(lambda row: brt_n_v1(row), axis=1)

# Rule 3. Jika resiko banjir bandang tinggi V3(n) ≥ 10
def brt_n_v3(row):
  if row['BBNDGTING'] < 10:
    return 0
  else:
    return row['BBNDGTING']

# Tambah kolom BRT(n) V3 berdasarkan hasil rule 3
dataset['BRT(n) V3'] = dataset.apply(lambda row: brt_n_v3(row), axis=1)

# Rule 4. Jika resiko tsunami tinggi V5(n) ≥ 5
def brt_n_v5(row):
  if row['TSUNTING'] < 5:
    return 0
  else:
    return row['TSUNTING']

# Tambah kolom BRT(n) V5 berdasarkan hasil rule 4
dataset['BRT(n) V5'] = dataset.apply(lambda row: brt_n_v5(row), axis=1)

# Rule 5. Jika resiko likuifaksi tinggi V7(n) ≥ 5
def brt_n_v7(row):
  if row['LIKTING'] < 5:
    return 0
  else:
    return row['LIKTING']

# Tambah kolom BRT(n) V7 berdasarkan hasil rule 5
dataset['BRT(n) V7'] = dataset.apply(lambda row: brt_n_v7(row), axis=1)

# Rule 6. Jika resiko gempa tinggi V9(n) ≥ 5
def brt_n_v9(row):
  if row['GEMTING'] < 10:
    return 0
  else:
    return row['GEMTING']

# Tambah kolom BRT(n) V9 berdasarkan hasil rule 6
dataset['BRT(n) V9'] = dataset.apply(lambda row: brt_n_v9(row), axis=1)

# Rule 7. Jika resiko longsor tinggi V11(n) ≥ 10
def brt_n_v11(row):
  if row['LONGTING'] < 10:
    return 0
  else:
    return row['LONGTING']

# Tambah kolom BRT(n) V11 berdasarkan hasil rule 7
dataset['BRT(n) V11'] = dataset.apply(lambda row: brt_n_v11(row), axis=1)

# Perhitungan skor kumulatif
def brt_n_all(row):
  return (
    row['BRT(n) V1']
    + row['BRT(n) V3']
    + row['BRT(n) V5']
    + row['BRT(n) V7']
    + row['BRT(n) V9']
    + row['BRT(n) V11']
  )

# Tambah kolom BRT(n) All berdasarkan hasil skor kumulatif
dataset['BRT(n) All'] = dataset.apply(lambda row: brt_n_all(row), axis=1)

# Lihat beberapa baris pertama, sekedar mengecek
# dataset.head()

# Simpan hasil skor ke file baru
dataset.to_csv('KRB_Grid_SB250_dengan_hasil_skor.csv')
```
